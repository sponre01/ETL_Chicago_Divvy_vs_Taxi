```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Extract Divvy Files
Divvy_Trips_2017_Q4_file = "Divvy_Trips_2017_Q4.csv"
Divvy_Trips_2017_Q3_file = "Divvy_Trips_2017_Q3.csv"
Divvy_Trips_2017_Q2_file = "Divvy_Trips_2017_Q2.csv"
Divvy_Trips_2017_Q1_file = "Divvy_Trips_2017_Q1.csv"
Divvy_Trips_2016_Q4_file = "Divvy_Trips_2016_Q4.csv"
Divvy_Trips_2016_Q3_file = "Divvy_Trips_2016_Q3.csv"
Divvy_Stations_2017_Q3Q4_file = "Divvy_Stations_2017_Q3Q4.csv"
Divvy_Stations_2017_Q1Q2_file = "Divvy_Stations_2017_Q3Q4.csv"
Divvy_Stations_2016_Q4_file = "Divvy_Stations_2016_Q4.csv"
Divvy_Stations_2016_Q3_file = "Divvy_Stations_2016_Q3.csv"
```

```python
# Read each Divvy Data File and store into Pandas DataFrame
Divvy_Trips_2017_Q4_df = pd.read_csv(Divvy_Trips_2017_Q4_file)
Divvy_Trips_2017_Q3_df = pd.read_csv(Divvy_Trips_2017_Q3_file)
Divvy_Trips_2017_Q2_df = pd.read_csv(Divvy_Trips_2017_Q2_file)
Divvy_Trips_2017_Q1_df = pd.read_csv(Divvy_Trips_2017_Q1_file)
Divvy_Trips_2016_Q4_df = pd.read_csv(Divvy_Trips_2016_Q4_file)
Divvy_Trips_2016_Q3_df = pd.read_csv(Divvy_Trips_2016_Q3_file)

Divvy_Stations_2017_Q3Q4_df = pd.read_csv(Divvy_Stations_2017_Q3Q4_file)
Divvy_Stations_2017_Q1Q2_df = pd.read_csv(Divvy_Stations_2017_Q1Q2_file)
Divvy_Stations_2016_Q4_df = pd.read_csv(Divvy_Stations_2016_Q4_file)
Divvy_Stations_2016_Q3_df = pd.read_csv(Divvy_Stations_2016_Q3_file)
```

```python
# Combine Trips Divvy Data File and store into Pandas DataFrame
Divvy_Trips_2017_Q3Q4_df = Divvy_Trips_2017_Q4_df.append(Divvy_Trips_2017_Q3_df)
Divvy_Trips_2017_Q2Q3Q4_df = Divvy_Trips_2017_Q3Q4_df.append(Divvy_Trips_2017_Q2_df)
Divvy_Trips_2017_df = Divvy_Trips_2017_Q2Q3Q4_df.append(Divvy_Trips_2017_Q1_df)
#Divvy_Trips_2017_df.head()
Divvy_Trips_2016_df = Divvy_Trips_2016_Q3_df.append(Divvy_Trips_2016_Q4_df)
Divvy_Trips_df = Divvy_Trips_2017_df.append(Divvy_Trips_2016_df)
Divvy_Trips_df.head()
```
