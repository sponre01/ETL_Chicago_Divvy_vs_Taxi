```python
#Establish client and data base to pull from using api_key :
client = Socrata("data.cityofchicago.org",api_key)

```

```python
# choose which columns to include in the later dataframe :
ct_sel = 'distinct dropoff_census_tract, dropoff_centroid_latitude, dropoff_centroid_longitude, dropoff_community_area'
```

```python
# get the data using ct_sel group with 5000 line limit :
results_census_tract = client.get("wrvz-psew", select=ct_sel, limit =5000)
```

```python
census_track_loc  = pd.DataFrame.from_records(results_census_tract)
census_track_loc = census_track_loc.dropna(axis=0)
```
